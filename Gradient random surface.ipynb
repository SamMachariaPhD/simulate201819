```python
import matplotlib as mpl
mpl.rc('text', usetex=True)
mpl.rc('font', family='serif')
mpl.rcParams['font.size'] = 14

xrnd = 30*np.random.uniform(size=1000)
yrnd = 30*np.random.uniform(size=1000)
plt.figure(figsize=(12,12))
plt.axes(aspect='equal')

plt.subplot(2,2,1)
plt.plot(xrnd,yrnd,'.',color='green',label='Myosin')
plt.plot([10,20],[15,15],'r-',lw=3, label='Actin')
plt.xlabel(r'Random x $(\mu m)$')
plt.ylabel(r'Random y $(\mu m)$')
plt.title('Random x,y myosin plot')
plt.legend(loc='upper left')

plt.subplot(2,2,2)
plt.hist(yrnd,bins=30, alpha=1, histtype='bar', ec='black',color='cyan', label='Y histogram')
plt.plot([0,30],[30,30],'b--',lw=3, label='Distribution')
plt.legend(loc='upper left')

plt.subplot(2,2,3)
plt.hist(xrnd,bins=30, alpha=1, histtype='bar', ec='black',color='green', label='X histogram')
plt.plot([0,30],[30,30],'b--',lw=3, label='Distribution')
plt.legend(loc='upper left')

plt.savefig('uniformRnd.svg',bbox_inches='tight',format='svg',dpi=500)
```

```python
import matplotlib as mpl
mpl.rc('text', usetex=True)
mpl.rc('font', family='serif')
mpl.rcParams['font.size'] = 14

xtrnd = np.random.triangular(0,0,30,size=1000)
ytrnd = np.random.triangular(0,0,30,size=1000)
plt.figure(figsize=(12,12))
plt.axes(aspect='equal')

plt.subplot(2,2,1)
plt.plot(xtrnd,yrnd,'.',color='green',label='Myosin')
plt.xlabel(r'Random x $(\mu m)$')
plt.ylabel(r'Random y $(\mu m)$')
plt.title('Random x,y myosin plot')
plt.legend(loc='upper left')

plt.subplot(2,2,2)
plt.hist(yrnd,bins=30,alpha=1,histtype='bar',ec='black',color='cyan', label='Y histogram')
plt.plot([0,30],[30,30],'b--',lw=3, label='Distribution')
plt.legend(loc='upper left')

plt.subplot(2,2,3)
plt.hist(xtrnd,bins=30,alpha=1,histtype='bar',ec='black',color='green', label='X histogram')
plt.plot([0,30],[60,0],'b--',lw=3, label='Distribution')
plt.legend(loc='upper left')

plt.savefig('triangularRnd.svg',bbox_inches='tight',format='svg',dpi=500)
```

```python
import matplotlib as mpl
mpl.rc('text', usetex=True)
mpl.rc('font', family='serif')
mpl.rcParams['font.size'] = 14

xRnd = np.random.triangular(0,0,100,size=6000)
yRnd = 100*np.random.uniform(size=6000)
plt.figure(figsize=(12,12))
plt.axes(aspect='equal')

plt.plot(xRnd,yRnd,'.',color='green',label='Myosin')
plt.plot([40,60],[50,50],'r-',lw=3, label='Actin')
plt.xlabel(r'Random x $(\mu m)$')
plt.ylabel(r'Random y $(\mu m)$')
plt.title('Random x,y myosin plot')
plt.legend(loc='upper left')

plt.savefig('triangularDist.svg',bbox_inches='tight',format='svg',dpi=500)
```
