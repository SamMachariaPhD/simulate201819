```python
Urnd = np.random.uniform(size=5000)
Brnd = np.random.binomial(5000,0.5,size=5000)
Trnd = np.random.triangular(0,0,20,size=5000)
Ernd = np.random.exponential(size=5000)
```

```python
plt.figure(figsize=(10,10))
plt.plot(Urnd,'.')
```

```python
plt.figure(figsize=(10,10))
plt.plot(Brnd,'.')
```

```python
plt.figure(figsize=(10,10))
plt.plot(Trnd,'.')
```

```python
plt.figure(figsize=(10,10))
plt.plot(Ernd,'.')
```
