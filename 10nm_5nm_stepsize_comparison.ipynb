```python
columns = ['r', 'b', 'atp', 'md', 'v_av', 'v_sd', 'v2_av', 'v2_sd', 'v3_av', 'v3_sd', 'v4_av', 'v4_sd']
#v = 10nm, v2 = 5 nm, v3 = 2.5 nm, v4 = 7.5 nm
df = pd.read_csv('bm_grinding_vel.csv', names=columns)

x=df['r']; y=df['v_av']; dev=df['v_sd']; y2=df['v2_av']; dev2=df['v2_sd']
y3=df['v3_av']; dev3=df['v3_sd'];y4=df['v4_av']; dev4=df['v4_sd']
plt.figure(figsize=(10,6), dpi=500)
plt.errorbar(x,y,yerr=dev, ecolor='b', capsize=7, \
             mec='blue', color='green', marker='D', linewidth=2, \
             markersize=7, label='10 nm speed deviation')
plt.errorbar(x,y2,yerr=dev2, ecolor='r', capsize=7, \
             mec='red', color='yellow', marker='D', linewidth=2, \
             markersize=7, label='5 nm speed deviation')
plt.errorbar(x,y3,yerr=dev3, ecolor='g', capsize=7, \
             mec='green', color='red', marker='D', linewidth=2, \
             markersize=7, label='2.5 nm speed deviation')
plt.errorbar(x,y4,yerr=dev4, ecolor='cyan', capsize=7, \
             mec='cyan', color='magenta', marker='D', linewidth=2, \
             markersize=7, label='7.5 nm speed deviation')
plt.xlabel('Active motor ratio', fontdict=font); plt.ylabel('Speed ($\mu m/sec$)', fontdict=font)
plt.title('Actin Filament Grinding Movement [ATP = 2000, MD = 3000]'); plt.legend(loc='upper left')#; plt.grid()
#plt.gca().set_aspect('equal', adjustable='box')
#plt.savefig('compr_actin_grinding.svg',bbox_inches='tight', format='svg')
#plt.savefig('compr_actin_grinding.png',bbox_inches='tight', format='png',dip=300)
plt.savefig('compr_actin_grinding.svg',bbox_inches='tight', format='svg',dip=300)
```
