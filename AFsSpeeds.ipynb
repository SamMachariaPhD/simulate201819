```python
plt.figure(figsize=(15,7))
plt.axis('tight')
plt.suptitle('ATP vs. Actin Filament Grinding Speed [MD = 3000]', fontsize=16)

plt.subplots_adjust(wspace=0.2)
plt.subplots_adjust(hspace=0.5)

plt.subplot(2,3,1) # ===================================================
x = atp; y = r10['speed']; sDev = r10['deviation']

plt.errorbar(x,y,yerr=sDev, ecolor='dodgerblue', capsize=3, \
             mec='purple', color='green', marker='o', linestyle='--', linewidth=1, \
             markersize=5, label='Simulation results')

xN = np.linspace(x.min(),x.max(),100)
sDevS = spline(x,sDev,xN)
ys = spline(x,y,xN)

plt.fill_between(xN,ys-sDevS,ys+sDevS,alpha=0.4,edgecolor='#CC4F18',facecolor='#FF9848', label='Speed deviation')

def func(x,km):
    return (x*y.max())/(x+km)
popt, pcov = curve_fit(func, x,y)

plt.plot(x,func(x,*popt),'g-', lw=2.5, label="Michaelis-Menten's fitting")

kmVal = popt[0]
yVal = (kmVal*y.max())/(kmVal+popt[0])
#plt.axhline(yVal, linestyle='--', color='gray', label='Average grinding speed')
#plt.text(2,yVal, str(round(yVal,2)), va='center', ha='center', backgroundcolor='w')

plt.xticks(np.arange(0,3300,step=600))
plt.yticks(np.arange(2,13,step=2))

plt.ylabel('Speed ($\mu m/sec$)', fontsize=14)
plt.title(r'Ratio = 1.0, $k_M\ =\ %0.2f,\ \frac{1}{2}v_{max}\ =\ %0.2f$'%(kmVal,yVal))
plt.legend(loc='lower right')
plt.grid(color='gray',which='both',linestyle=':',linewidth=0.25)

plt.subplot(2,3,2) # ===================================================
x = atp; y = r09['speed']; sDev = r09['deviation']

plt.errorbar(x,y,yerr=sDev, ecolor='dodgerblue', capsize=3, \
             mec='purple', color='green', marker='o', linestyle='--', linewidth=1, \
             markersize=5, label='Simulation results')

xN = np.linspace(x.min(),x.max(),100)
sDevS = spline(x,sDev,xN)
ys = spline(x,y,xN)

plt.fill_between(xN,ys-sDevS,ys+sDevS,alpha=0.4,edgecolor='#CC4F18',facecolor='#FF9848', label='Speed deviation')

def func(x,km):
    return (x*y.max())/(x+km)
popt, pcov = curve_fit(func, x,y)

plt.plot(x,func(x,*popt),'g-', lw=2.5, label="Michaelis-Menten's fitting")

kmVal = popt[0]
yVal = (kmVal*y.max())/(kmVal+popt[0])

plt.xticks(np.arange(0,3300,step=600))
plt.yticks(np.arange(2,13,step=2))

plt.title(r'Ratio = 0.9, $k_M\ =\ %0.2f,\ \frac{1}{2}v_{max}\ =\ %0.2f$'%(kmVal,yVal))
#plt.legend(loc='lower right')
plt.grid(color='gray',which='both',linestyle=':',linewidth=0.25)

plt.subplot(2,3,3) # ===================================================
x = atp; y = r08['speed']; sDev = r08['deviation']

plt.errorbar(x,y,yerr=sDev, ecolor='dodgerblue', capsize=3, \
             mec='purple', color='green', marker='o', linestyle='--', linewidth=1, \
             markersize=5, label='Simulation results')

xN = np.linspace(x.min(),x.max(),100)
sDevS = spline(x,sDev,xN)
ys = spline(x,y,xN)

plt.fill_between(xN,ys-sDevS,ys+sDevS,alpha=0.4,edgecolor='#CC4F18',facecolor='#FF9848', label='Speed deviation')

def func(x,km):
    return (x*y.max())/(x+km)
popt, pcov = curve_fit(func, x,y)

plt.plot(x,func(x,*popt),'g-', lw=2.5, label="Michaelis-Menten's fitting")

kmVal = popt[0]
yVal = (kmVal*y.max())/(kmVal+popt[0])

plt.xticks(np.arange(0,3300,step=600))
plt.yticks(np.arange(2,13,step=2))

plt.title(r'Ratio = 0.8, $k_M\ =\ %0.2f,\ \frac{1}{2}v_{max}\ =\ %0.2f$'%(kmVal,yVal))
#plt.legend(loc='lower right')
plt.grid(color='gray',which='both',linestyle=':',linewidth=0.25)

plt.subplot(2,3,4) # ===================================================
x = atp; y = r07['speed']; sDev = r07['deviation']

plt.errorbar(x,y,yerr=sDev, ecolor='dodgerblue', capsize=3, \
             mec='purple', color='green', marker='o', linestyle='--', linewidth=1, \
             markersize=5, label='Simulation results')

xN = np.linspace(x.min(),x.max(),100)
sDevS = spline(x,sDev,xN)
ys = spline(x,y,xN)

plt.fill_between(xN,ys-sDevS,ys+sDevS,alpha=0.4,edgecolor='#CC4F18',facecolor='#FF9848', label='Speed deviation')

def func(x,km):
    return (x*y.max())/(x+km)
popt, pcov = curve_fit(func, x,y)

plt.plot(x,func(x,*popt),'g-', lw=2.5, label="Michaelis-Menten's fitting")

kmVal = popt[0]
yVal = (kmVal*y.max())/(kmVal+popt[0])

plt.xticks(np.arange(0,3300,step=600))
plt.yticks(np.arange(2,13,step=2))

plt.xlabel('ATP', fontsize=14); plt.ylabel('Speed ($\mu m/sec$)', fontsize=14)
plt.title(r'Ratio = 0.7, $k_M\ =\ %0.2f,\ \frac{1}{2}v_{max}\ =\ %0.2f$'%(kmVal,yVal))
#plt.legend(loc='lower right')
plt.grid(color='gray',which='both',linestyle=':',linewidth=0.25)

plt.subplot(2,3,5) # ===================================================
x = atp; y = r06['speed']; sDev = r06['deviation']

plt.errorbar(x,y,yerr=sDev, ecolor='dodgerblue', capsize=3, \
             mec='purple', color='green', marker='o', linestyle='--', linewidth=1, \
             markersize=5, label='Simulation results')

xN = np.linspace(x.min(),x.max(),100)
sDevS = spline(x,sDev,xN)
ys = spline(x,y,xN)

plt.fill_between(xN,ys-sDevS,ys+sDevS,alpha=0.4,edgecolor='#CC4F18',facecolor='#FF9848', label='Speed deviation')

def func(x,km):
    return (x*y.max())/(x+km)
popt, pcov = curve_fit(func, x,y)

plt.plot(x,func(x,*popt),'g-', lw=2.5, label="Michaelis-Menten's fitting")

kmVal = popt[0]
yVal = (kmVal*y.max())/(kmVal+popt[0])

plt.xticks(np.arange(0,3300,step=600))
plt.yticks(np.arange(2,13,step=2))

plt.xlabel('ATP', fontsize=14)
plt.title(r'Ratio = 0.6, $k_M\ =\ %0.2f,\ \frac{1}{2}v_{max}\ =\ %0.2f$'%(kmVal,yVal))
#plt.legend(loc='lower right')
plt.grid(color='gray',which='both',linestyle=':',linewidth=0.25)

plt.subplot(2,3,6) # ===================================================
x = atp; y = r05['speed']; sDev = r05['deviation']

plt.errorbar(x,y,yerr=sDev, ecolor='dodgerblue', capsize=3, \
             mec='purple', color='green', marker='o', linestyle='--', linewidth=1, \
             markersize=5, label='Simulation results')

xN = np.linspace(x.min(),x.max(),100)
sDevS = spline(x,sDev,xN)
ys = spline(x,y,xN)

plt.fill_between(xN,ys-sDevS,ys+sDevS,alpha=0.4,edgecolor='#CC4F18',facecolor='#FF9848', label='Speed deviation')

def func(x,km):
    return (x*y.max())/(x+km)
popt, pcov = curve_fit(func, x,y)

plt.plot(x,func(x,*popt),'g-', lw=2.5, label="Michaelis-Menten's fitting")

kmVal = popt[0]
yVal = (kmVal*y.max())/(kmVal+popt[0])

plt.xticks(np.arange(0,3300,step=600))
plt.yticks(np.arange(2,13,step=2))

plt.xlabel('ATP', fontsize=14)
plt.title(r'Ratio = 0.5, $k_M\ =\ %0.2f,\ \frac{1}{2}v_{max}\ =\ %0.2f$'%(kmVal,yVal))
#plt.legend(loc='lower right')
plt.subplots_adjust(hspace=0.25)
plt.grid(color='gray',which='both',linestyle=':',linewidth=0.25)
plt.savefig('speedATP.svg',fmt='svg',dip=3000,bbox_inches='tight')
```
