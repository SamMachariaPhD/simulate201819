```python
# prepared by Sam. -- Prof. Nitta Lab.
import matplotlib.pyplot as plt; import pandas as pd; import numpy as np
from scipy.optimize import curve_fit
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 14,
        }
%matplotlib inline
```

```python
columns = ['r', 'b', 'atp', 'md', 'v_av', 'v_sd']
df = pd.read_csv('vel_changes.csv', names=columns)

r_change = df[0:10]
x = r_change['r']
y = r_change['v_av']
```

```python
plt.figure(figsize=(14,10), dpi=500)
colors=['green','green','white','blue','white','green','green','green','green','green']
plt.errorbar(x,y,yerr=r_change['v_sd'], ecolor='blue', capsize=7, \
             mec='blue', color='green', marker='D', linewidth=2, \
             markersize=7, label='Mean vel. deviation (ATP = 2000, MD = 3000)')

plt.errorbar(x[2:5],y[2:5], capsize=7, mec='blue', color='yellow', \
             marker='D', linewidth=2, markersize=7, label='Detached filament')

plt.xlabel('Specie Ratio', fontdict=font); plt.ylabel('Mean Velocity', fontdict=font)
plt.title('Actin Filament Movement'); plt.legend(loc='upper left'); plt.grid()
```
