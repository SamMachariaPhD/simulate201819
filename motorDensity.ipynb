```python
plt.figure(figsize=(15,6))
plt.axis('tight')
plt.suptitle('Binding Motor Number vs. Probability Density [ATP = 2000, R = 1.0, Step size = 10 nm, Filament 1]', fontsize=16)

plt.subplots_adjust(wspace=0.7)
plt.subplots_adjust(hspace=0.4)
plt.axes().set_aspect('equal')

byn = 20; md = 1000


#(mu,sigma)=norm.fit(df6)
#n,bins,patches = plt.hist(df6,byn, density=True, alpha=1,histtype='bar',ec='black',color='green')
#plt.plot(bins,1/(sigma*np.sqrt(2*np.pi))*np.exp(-(bins-mu)**2/(2*sigma**2)),'b-',lw=1.3)


for i in range(1,11):
    plt.subplot(2,5,i)
    (mu,sigma)=norm.fit(bm873[i])
    n,bins,patches = plt.hist(bm873[i],byn, density=True, alpha=1,histtype='bar',ec='black',color='green')
    #(mu,sigma)=(np.round(bm873[i].mean(),0), np.round(np.std(bm873[i]),0))
    #plt.hist(bm873[i],byn, density=True, alpha=1,histtype='bar',ec='black',color='green')
    plt.title(r'MD = %d, $\mu=%.2f,\ \sigma=%.2f$' %(md, mu, sigma))
    plt.grid(color='gray',which='both',linestyle=':',linewidth=0.25)
    if i>2:
        plt.plot(bins,1/(sigma*np.sqrt(2*np.pi))*np.exp(-(bins-mu)**2/(2*sigma**2)),'b-',lw=1.3)
        plt.yticks(np.arange(0,0.17,step=0.04), fontsize=14)
        plt.xticks(np.arange(0,81,step=20))
    elif i<3:
        plt.yticks(np.arange(0,2.1,step=0.4), fontsize=14)
        plt.xticks(np.arange(0,81,step=20))
    if i==1 or i==6:
        plt.ylabel('Density', fontsize=14)
    if i>5:
        plt.xlabel('Motor number', fontsize=14)
    md = md+500
plt.savefig('bmHist873.svg',fmt='svg',dpi=2000, bbox_inches='tight')
```

```python
AF873 = []; md = 500

for j in range(11):
    AF873MD = pd.read_csv('TipXY'+str(md)+'seed873.txt', names=['t', 'x', 'y'], delim_whitespace=True)
    AF873MD = AF873MD.drop(['t'], axis=1) # we don't need 't'. NB: dt = 0.01*1 = 0.01 sec
    AF873MD = AF873MD.iloc[0::10, :] #pick every 9th row starting from the first. Here dt = 0.01*10 = 0.1
    AF873.append(AF873MD)
    md = md+500
```

```python
plt.figure(figsize=(15,6))
plt.suptitle('Leading Tip Trajectory [Filament 1]', fontsize=16)
plt.subplots_adjust(wspace=0.4)
plt.subplots_adjust(hspace=0.4)
plt.axes().set_aspect('equal')

md = 1000

for i in range(1,11):
    plt.subplot(2,5,i)
    plt.plot(AF873[i]['x'],AF873[i]['y'], label='Leading tip', color='red', marker='.', linestyle='-', linewidth=0.5, markersize=7)
    plt.xticks(np.arange(-15,26,step=10),fontsize=14)
    plt.yticks(np.arange(-15,26,step=10),fontsize=14)
    plt.title('MD = %d'%(md), fontsize=14)
    if i==1 or i==6:
        plt.ylabel('Y ($\mu m$)', fontsize=16)
    if i>5:
        plt.xlabel('X ($\mu m$)', fontsize=16)
    md = md+500
plt.savefig('traj873.svg',fmt='svg',dpi=2000, bbox_inches='tight')
```
